```python
import numpy as np
import matplotlib.pyplot as plt

# Definir señales de ejemplo
x = np.array([1, 2, 3, 4, 5])
h = np.array([1, -1, 1])

# Realizar convolución
y = np.convolve(x, h)

# Crear figura con dos subplots
fig, axs = plt.subplots(2, 1)

# Primer subplot: señal de entrada y señal de respuesta al impulso
axs[0].stem(x, linefmt='C0-', markerfmt='C0o', label='Señal de entrada')
axs[0].stem(h, linefmt='C1-', markerfmt='C1x', label='Respuesta al impulso')
axs[0].legend()
axs[0].set_xlabel('Muestras')
axs[0].set_ylabel('Amplitud')
axs[0].set_title('Señal de entrada y respuesta al impulso')

# Segundo subplot: señal resultante de la convolución
axs[1].stem(y, linefmt='C2-', markerfmt='C2^', label='Señal resultante')
axs[1].legend()
axs[1].set_xlabel('Muestras')
axs[1].set_ylabel('Amplitud')
axs[1].set_title('Señal resultante de la convolución')

# Mostrar la figura
fig.set_size_inches(10, 6)
plt.tight_layout()
plt.show()
```
