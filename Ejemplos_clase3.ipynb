#Representacion de Secuencias

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
import random

x_n=[random.random() for i in range(10)]
print("Secuencia de datos:","\n",x_n)
plt.stem(x_n, use_line_collection=True)

# Configuración de la gráfica
plt.xlabel('Numero de muestra')
plt.ylabel('Amplitud')
plt.title('Secuencia de datos')
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

# Graficas de escalon, delta, rampa y exponencial.



```python
import numpy as np
import matplotlib.pyplot as plt

# Señal escalón
x1 = np.ones(10)
plt.stem(x1)
plt.title('Señal escalón')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.grid()
plt.show()

# Señal delta
x2 = np.zeros(10)
x2[5] = 1
plt.stem(x2)
plt.title('Señal delta')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.grid()
plt.show()

# Señal rampa
x3 = np.arange(10)
plt.stem(x3)
plt.title('Señal rampa')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.grid()
plt.show()

# Señal exponencial
n = np.arange(10)
alpha = 0.5
x4 = alpha**n
plt.stem(x4)
plt.title('Señal exponencial')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```
