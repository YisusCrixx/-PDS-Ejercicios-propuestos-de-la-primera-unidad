Ejemplo de transformada z

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

#definimos la seal que queremos transformar en funcin del tiempo. Para este ejemplo, definimos una seal senoidal:
t = np.linspace(0, 1, 1000, endpoint=False)
x = np.sin(2*np.pi*10*t)

#la funcin "signal.zpk2tf" de la librera "scipy" para convertir los polos y ceros de la seal en su forma de coeficientes de un polinomio. En este ejemplo, se utiliza una funcin de transferencia de segundo orden con un polo y un cero en z=0.9:
b, a = signal.zpk2tf([0.9], [0.1], 1)

#utilizamos la funcin "signal.lfilter" de la librera "scipy" para realizar la transformada Z de la seal utilizando los coeficientes del polinomio
zi = signal.lfilter_zi(b, a)
z, _ = signal.lfilter(b, a, x, zi=zi*x[0])
#Graficamos la seal original
fig, ax = plt.subplots(2, 1, figsize=(8, 6))
ax[0].plot(t, x)
ax[0].set_title('Seal Original')
ax[0].set_xlabel('Tiempo [s]')
ax[0].set_ylabel('Amplitud')
#Graficamos la transformada z
ax[1].plot(np.arange(len(z)), np.abs(z))
ax[1].set_title('Transformada Z')
ax[1].set_xlabel('ndice')
ax[1].set_ylabel('Amplitud')
plt.tight_layout()
plt.show()

```

La transformada Z en la parte inferior, los puntos en el eje horizontal corresponden al ndice de la muestra y la amplitud se muestra en el eje vertical.
