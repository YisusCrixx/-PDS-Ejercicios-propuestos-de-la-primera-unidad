```python
import numpy as np
import matplotlib.pyplot as plt

# Defininimos una señal periódica
frecuencia = 5   # Frecuencia en Hz
periodo = 1 / frecuencia   # Periodo en segundos
tiempo = np.linspace(0, 2*periodo, 1000)  # Vector de tiempo
señal = np.sin(2*np.pi*frecuencia*tiempo)  # Señal senoidal

# Muestrear la señal
frecuencia_muestreo = 50   # Frecuencia de muestreo en Hz
periodo_muestreo = 1 / frecuencia_muestreo   # Periodo de muestreo en segundos
tiempo_muestreo = np.arange(0, tiempo[-1], periodo_muestreo)  # Vector de tiempo de muestreo
señal_muestreada = np.sin(2*np.pi*frecuencia*tiempo_muestreo)  # Señal muestreada

# Visualizar las señales
plt.plot(tiempo, señal,label='sin(x)')
plt.title('Señal Continua y Muestreada')
plt.stem(tiempo_muestreo, señal_muestreada,label='sin(x) muestreada',linefmt='r-', markerfmt='go')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```
