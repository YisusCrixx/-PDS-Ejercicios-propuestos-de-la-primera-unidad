#Ejemplo de señal reflexionada y de filtro de primera diferencia

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear una señal discreta de ejemplo
x = np.array([1, 2, 3, 4, 5, 6])

# Reflexionar la señal
y = np.flip(x)

# Aplicar filtro de primera diferencia
d = np.diff(x)

# Crear figura con dos subplots
fig, axs = plt.subplots(2, 1)

# Primer subplot: señal original y señal reflexionada
axs[0].stem(x, linefmt='C0-', markerfmt='C0o', label='Original')
axs[0].stem(y, linefmt='C1-', markerfmt='C1x', label='Reflexionada')
axs[0].legend()
axs[0].set_xlabel('Muestras')
axs[0].set_ylabel('Amplitud')
axs[0].set_title('Señal original y señal reflexionada')

# Segundo subplot: señal original y señal filtrada
axs[1].stem(x, linefmt='C0-', markerfmt='C0o', label='Original')
axs[1].stem(d, linefmt='C2-', markerfmt='C2^', label='Filtrada')
axs[1].legend()
axs[1].set_xlabel('Muestras')
axs[1].set_ylabel('Amplitud')
axs[1].set_title('Señal original y con filtro de primera diferencia')

# Mostrar la figura
fig.set_size_inches(12, 8)
plt.tight_layout()
plt.show()
```

Se puede observar cómo la señal filtrada (correspondiente a la primera diferencia) tiene un valor nulo en la primera muestra, ya que la diferencia no se puede calcular para la primera muestra de la señal original. En general, el filtro de primera diferencia puede utilizarse para calcular la tasa de cambio o la derivada de una señal discreta.
