```python
#   EJEMPLO DE NYQUIST Y ALIASING:
import numpy as np
import matplotlib.pyplot as plt # libreria usada para graficas
f=1.0 # Frecuencia de la senal
fs4=22.0 # Frecuencia de muestreo 22Hz
fs5=2*f #nyquist
t3=np.arange(0, 2.0, 1.0/fs4) # Vector de tiempo
t4=np.arange(0, 2.0, 1.0/fs5) # Vector de tiempo
x1 = np.sin(2*np.pi*f*t3)
x2= np.sin(2*np.pi*f*t4)
plt.figure(figsize=(10,6))
plt.plot(t3,x1,'r', label='Seno')
plt.plot(t4,x2,'bs',label='Seno muestreado nyquist')
plt.legend(loc ="upper right")
plt.xlabel('Time',fontsize=18)
plt.ylabel('Amplitude',fontsize=18)
plt.show()
```

La señal senoidal se crea con una frecuencia de 1 Hz y luego se muestrea con una frecuencia de muestreo de 22 Hz, que está por encima de la frecuencia de Nyquist (2 Hz en este caso). Luego, la misma señal se muestrea con la frecuencia de Nyquist de 2 Hz, lo que resulta en una señal discreta que no representa correctamente la señal original, dandose el fenomeno del aliasing

```python
fs6=(7.0 + 1.0) * 22.0 # Frecuencia de muestreo 22Hz
f1= (7.0 + 1.0)  # Frecuencia de la senal
f2=2.0 * fs6 + (7.0 + 2.0)

t=np.arange(0, 2.0, 1.0/fs6) # Vector de tiempo
x1 = np.sin(2*np.pi*f1*t)
x2 = np.sin(2*np.pi*f2*t)
plt.figure(figsize=(10,6))
plt.plot(t,x1,'r', label='Seno')
plt.plot(t,x2,'b',label='Seno ')
plt.legend(loc ="upper right")
plt.xlabel('Time',fontsize=18)
plt.ylabel('Amplitude',fontsize=18)
plt.show()
```

Se crea una señal que contiene dos senos, uno con una frecuencia de 8 Hz y otro con una frecuencia de 222 Hz. La frecuencia de muestreo utilizada es de 198 Hz, que es inferior a la frecuencia de Nyquist de la señal (que sería de 444 Hz en este caso), lo que produce el fenómeno de aliasing. El aliasing hace que la señal resultante tenga una frecuencia aparente de 16 Hz en lugar de 8 Hz, lo que se puede observar en la gráfica. Además, la señal de alta frecuencia de 222 Hz no puede ser representada correctamente debido a la limitación en la frecuencia de muestreo, lo que resulta en una forma de onda que no se asemeja a la señal original.
