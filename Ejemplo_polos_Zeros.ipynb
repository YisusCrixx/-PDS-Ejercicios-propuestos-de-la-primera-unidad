Ejemplo de polos y zeros

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de transferencia
num = np.array([1, 0, 2])
den = np.array([1, -0.5, 0.25])

# Calcular los polos y ceros de la función de transferencia
zeros = np.roots(num)
poles = np.roots(den)

# Graficar los polos y ceros en el plano complejo
fig, ax = plt.subplots()
plt.plot(np.real(zeros), np.imag(zeros), 'o', markersize=10, label='Ceros')
plt.plot(np.real(poles), np.imag(poles), 'x', markersize=10, label='Polos')
plt.axis('equal')
plt.legend(loc='best')
plt.title('Polos y Ceros en el Plano Complejo')
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginaria')
plt.show()
```
